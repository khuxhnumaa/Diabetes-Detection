```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df= pd.read_csv('file:///Users/khushnumaparveen/Downloads/diabetes.csv')
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
#to get statistical measure of dataset
print(df.describe())
```

```python
print(df['Outcome'].value_counts())
print("0-> diabetic\n1-> Non-diabetic")
```

```python
X= df.drop('Outcome',axis=1)
y= df['Outcome']
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_new= scaler.transform(X)
```

```python
X= X_new
y= df['Outcome']
```

```python
#splitting the dataset
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y,random_state=66)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python

sm= KNeighborsClassifier(n_neighbors=3)
```

```python
sm.fit(X_train,y_train)
```

```python
#training data prediction
X_train_pred= sm.predict(X_train)
print("Training set accuracy: {:.2f}".format(accuracy_score(X_train_pred,y_train)))
```

```python
#test dataset prediction
y_pred= sm.predict(X_test)
print("Model accuracy for test dataset: {:.2f}".format(accuracy_score(y_test,y_pred)))
```

```python
print("Classification Report\n {}".format(classification_report(y_test,y_pred)))
```

```python
print("Classification Report\n {}".format(classification_report(X_train_pred,y_train)))
```

```python
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print confusion matrix
print("Confusion Matrix:\n", cm)
```
